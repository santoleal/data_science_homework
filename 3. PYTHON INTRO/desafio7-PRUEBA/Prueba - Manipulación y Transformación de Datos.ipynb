```python
## Desde Anaconda Navigator se ha creado un entorno virtual, y desde terminal se ha instalado a través de pip install:
#  - Numpy, Pandas, Psycopg2-binary y Sqlalchemy
# Se adjunta en Github el archivo requirements.txt


# Se importan librerías a utilizar
import pandas as pd
import numpy as np
import psycopg2
import sqlalchemy
```

```python
# se establece string de conexión con Postgres

db_url ="postgresql://postgres:postgres@localhost:5432/classicmodels"

# se estrablece un objeto de conexión

from sqlalchemy import create_engine
engine = create_engine(db_url)
```

```python
# Se chequea la lectura de la bbdd

query_order = "SELECT * FROM orders"

df_order = pd.read_sql(query_order, engine)

df_order.head()
```

```python
# Se chequean nombres de tablas de la bbdd "classicmodels"

query_listado_tablas = "SELECT table_name FROM information_schema.tables WHERE table_schema = 'public';"

lista_tablas = pd.read_sql(query_listado_tablas, engine)

lista_tablas
```

```python
# Comprobada la conexión y lectura del sql, y conocidos sus tablas, se crea una función para cumplir con lo solicitado

def leer_tabla(nombre_tabla, engine):
    query_general = f"SELECT * FROM {nombre_tabla}"
    return pd.read_sql(query_general, engine)        
```

```python
# ahora se llama a la función y return se almacena en dataframes independientes, de acuerdo a cada tabla solicitada

# Orders (1/5)
df_orders = leer_tabla("orders", engine)

df_orders.head()

```

```python
# orderdetails (2/5)
df_orderdetails = leer_tabla("orderdetails", engine)

df_orderdetails.head()
```

```python
# customers (3/5)
df_customers = leer_tabla("customers", engine)

df_customers.columns
```

```python
# products (4/5)
df_products = leer_tabla("products", engine)

df_products.head()
```

```python
# employees (5/5)
df_employees = leer_tabla("employees", engine)

df_employees
```

#### 2. Realiza el cruce entre los DataFrames, asegurándote de utilizar correctamente el parámetro validate para asegurar la integridad referencial.

```python
# Primer cruce: df_orders y df_orderdetails (1/4)
df_unido = pd.merge(df_orders,df_orderdetails,on='orderNumber', how='outer', validate='one_to_many')

df_unido.info()
```

```python
# Segundo cruce: df_unido y df_customers (2/4)

df_unido = pd.merge(df_unido, df_customers, on='customerNumber', how='outer', validate= 'many_to_one')

df_unido.info()
```

```python
# Tercer cruce: df_unido y df_products (3/4)

df_unido = pd.merge(df_unido, df_products, on='productCode', how='outer', validate= 'many_to_one')

df_unido.info()
```

```python
# Cuarto cruce: df_unido y df_employees (4/4)

df_unido = pd.merge(df_unido, df_employees, left_on='salesRepEmployeeNumber', right_on='employeeNumber', validate= 'many_to_one')

df_unido.info()
```

```python
# chequeo final

df_unido.head()
```

#### 6. ¿Existen clientes que aún no han hecho ninguna compra? ¿Cuántos son?

```python
# Genero una pivot table para cruzar Customer Number con Venta, lo que indicará si compraron o no (en base al valor de venta)

pivote_ventas_cliente = pd.pivot_table(df_unido, index='customerNumber', values='venta', aggfunc='sum',  margins=True )

pivote_ventas_cliente


```

```python
# Se ordena valores

pivote_ventas_cliente = pivote_ventas_cliente.sort_values(by='venta', ascending=True)

pivote_ventas_cliente
```

```python
# Se crea una columna para identificar: quiénes han comprado /// quiénes no han comprado

# umbral para identificar las opciones
indicador_sin_compra = 0


# columna con opciones
pivote_ventas_cliente['Estatus'] = np.where(pivote_ventas_cliente['venta'] <= indicador_sin_compra, 'No ha comprado', 'Comprador')

# Se ordena 
pivote_ventas_cliente = pivote_ventas_cliente.sort_values(by='Estatus', ascending=False)

# Agrego índices para contabilizar mejor
pivote_ventas_cliente = pivote_ventas_cliente.reset_index()
pivote_ventas_cliente
```
